```python
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies.head().T
```

```python
movies.isnull().sum()
```

Features Selection

```python
movies= movies[['id', 'title', 'genres', 'overview']]
```

```python
movies['tags']= movies['overview']+movies['genres']
```

```python
movies
```

```python
new_data= movies.drop(columns=['genres','overview'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=4803, stop_words='english')
```

```python
vector= cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title']=="John Carter"].index[0]
```

```python
distance= sorted(list(enumerate(similarity[4])), reverse=True, key=lambda vector:vector[1])

for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommend(movies):
    index= new_data[new_data['title']==movies].index[0]
    distance= sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
    
```

```python
recommend('Iron Man')
```
